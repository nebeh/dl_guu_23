```python
import numpy as np
```

### Создание массива

```python
t = np.array([[1, 0], [2, 1]])
t1 = np.array([3, 5])
t2 = np.array([[-1, -2]])
print(t)
print(t1)
print(t2)
```

```python
Z = np.zeros((2,2,6))
O = np.ones((3,2,4,2))
E = np.eye(4)

print(Z)
print(E)
```

### Случайные числа

```python
rn = np.random.randn(3,5,4) # нормальное распределение
ru = np.random.rand(3,4) # равномерное распределение на отрезке [0,1]
print(ru)
```

### Форма и размер массива

```python
print(rn.shape) # размеры по осям
print(rn.size) # количество элементов в массиве
print(rn.ndim) # размерность массива
```

### Объединение массивов

```python
t = np.block([[t],[t1],[t2]]) # блочная матрица
print(t)
print(np.concatenate((t,t2),0)) # конкатенация по нулевой размерности
print(np.r_[t,t2]) # конкатенация по строкам
```

### Достуа к элементам массива

```python
print(t[1,0]) # 1-я строка, нулевой столбец
t[:,1] # все строки, 1-й столбец
t[-1,:] # все стобцы, последняя строка
t[-2:,:] # строки от предпоследней до последней, все стобцы
t[[0,2,3],:] # строки 0,2,3, стобцы все
t[::-1,:] #flip
print(t[0:4:2,:]) # строки с 0-го по 4-й с шагом 2, все колонки
t[::2,:] # строки с 0-го по последний с шагом 2, все колонки
```

### Логическое индексирование

```python
x = np.random.rand(1,5)
b = x>0.5
print(x)
print(b)
print(x[b])
print(x[x>0.5])
```

```python
A = np.random.rand(4,3)
B = A[A[:,0]>0.5,:] # оставляем в матрице B только те строки, для которых 1-я колонка > 0.5
print(A)
print(B)
```

### Поэлементные операции

```python
x = np.array([1,2,3])
y = np.array([-0.2,1.1,2])
print(x*y)
print(y**x)
A = np.random.rand(2,2)
B = np.random.rand(2,2)
print(A*B)
```

### Линейная алгебра

```python
print(A@B) # матриченое умножение
print(A.dot(B)) # матриченое умножение
```

```python
b = np.array([0,1])
print(A@b) # умножение матрицы на вектор
```

### 3D plots 

```python
t = np.linspace(-2,2)
x,y = np.meshgrid(t,t)
z = np.sin(x) + np.cos(y)

import matplotlib.pyplot as plt
fig = plt.figure(figsize=(16, 6))
ax = fig.add_subplot(121, projection='3d')
ax.plot_surface(x, y, z)
ax = fig.add_subplot(122)
ax.contourf(x, y, z)
```
